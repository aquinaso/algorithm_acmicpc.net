```python
def system_of_equations(a,b,c,d,e,f):
    (a*x1) + (b*y1) == c
    (d*x2) + (e*y2) == f

    if x1 == x2 and y1==y2:
        return print(f"{x1} \t {y1}")
    else:
        return print("뭔가 잘못됐어요")
```
